```python
liste = [[3,4.0,2,8.4,6],[0,2,0.2,4,6],[9,3.5,0.32,5,4]]
int_list = []
float_list = []
for i in liste:
    for j in i:
        if type(j) == int:
                int_list.append(j)
        elif type(j) == float:
                float_list.append(j)
print("int say覺lar", int_list)
print("float say覺lar", float_list)
                
        
```
